# Data Visualisation des articles

## Importation des libraires nécessaires

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
```

## Importation des données textuelles

```python
morocco_world_news = pd.read_csv('morocco_world_news_articles.csv')
```

```python
morocco_world_news_text = morocco_world_news['content']
```

```python
le_matin_articles = pd.read_csv('le_matin_articles.csv')
```

```python
le_matin_articles_text = le_matin_articles['content']
```

## Choix des mots à exclure

```python
stop_words = ['the', 'of', 'in', 'he', 'their', 'said', 'that', 'on', 'has', 'for', 'are', 'it', 'is', 'to', 'd', 'du', 'de', 'la', 'des', 'le', 'et', 'est', 'elle', 'une', 'en', 'que', 'aux', 'qui', 'ces', 'les', 'dans', 'sur', 'l', 'un', 'pour', 'par', 'il', 'ou', 'à', 'ce', 'a', 'sont', 'cas', 'plus', 'leur', 'se', 's', 'vous', 'au', 'c', 'aussi', 'toutes', 'autre', 'comme', 'or', 'and', 'then', 'if', 'with', 'when', 'who', 'why', 'else']
```

## Génération des nuages de mots

```python
morocco_world_news_wordcloud = WordCloud(background_color = 'white', stopwords = stop_words, max_words = 50).generate(morocco_world_news_text[90])
plt.imshow(morocco_world_news_wordcloud)
plt.axis("off")
plt.show();
```

```python
morocco_world_news_wordcloud.to_file('article_1.png')
```

```python
morocco_world_news_wordcloud = WordCloud(background_color = 'white', stopwords = stop_words, max_words = 50).generate(morocco_world_news_text[1000])
plt.imshow(morocco_world_news_wordcloud)
plt.axis("off")
plt.show();
```
